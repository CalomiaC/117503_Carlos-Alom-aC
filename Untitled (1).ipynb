```python
#5
V='AEIOU'

T='ÁÉÍÓÚáéíóú'

E=' '

T='ÁÉÍÓÚáéíóú'

D='0123456789'

Palabras_reservadas = ['and','del','for','is','raise','assert','if','else',\
'elif','from','lambda','return','break','global','not','try','class','except',\
'or','while','continue','exec','import','yield','def','finally','in','print']

#Contadores
cont_voc = 0
cont_til = 0
cont_dig = 0
cont_esp = 0
cont_res = 0

#Ingresamos la cadena de texto
cadena = input('Ingrese una cadena de texto: ')

#Vamos a devolver la cadena de texto a una lista para poder efectuar los 
#operadores
for i in range (len(cadena)):
    if cadena[i] in V:
        cont_voc += 1
    elif cadena[i] in T:
        cont_til += 1
    elif cadena[i] in D:
        cont_dig += 1
    elif cadena[i] in E:
        cont_esp += 1
    elif cadena[i] in Palabras_reservadas :
        cont_res += 1
shi = cadena.split() #transformamos a lista

for n, i in enumerate(shi): #comparamos
    if i in Palabras_reservadas:#Con las palabras reservadas. Si cumple se suma
        cont_res+=1        #al contador 
        
print(f'Se entraron {cont_voc} letras vocales en mayúscula')
print(f'Se entraron {cont_til} letras con tilde')
print(f'Se entraron {cont_dig} digitos')
print(f'Se entraron {cont_esp} espacios')
print(f'Se entraron {cont_res} palabras reservadas')
```

```python
#9
A = int(input('Digame el valor: '))
B = int(input('Dígame cuántos términos quieres: '))
vec = [A]
    
if A % 2 == 0 and B > 0:  
    for i in range(B-1):
        A = int(A / 2)
        vec.append(A)
    print(vec)
elif A % 2 != 0 and B > 0:
    for i in range(B-1):
        A = int(3*A + 1)
        vec.append(A)
    print(vec)
else:
    print('Error, ingrese los datos nuevamente')
```

```python
#12
ancho = input('Ancho del triangulo: ') #Ingresamos Ancho del triángulo
def triangulo(an):
    r = 0
    for i in range(int(an)-1,-1,-1):
        if i==int(an)-1 or i==int(an)-2 or i == 0 :
            print(' '*i,' *'*(int(an)-i))
        else:
            print(' '*i,' *',' '*r+'  *')
            r+=2
triangulo(ancho)
```

```python
#13
primero =input('Años bisiestos desde el año: ')
segundo = input('hasta el año: ')
def bisiestos(inicio,fin):
    bis=[]
    for i in range(int(inicio),int(fin)+1):
        if i%4 ==0:
            bis.append(i)
    if len(bis)>0:
        print('Hay '+ str(len(bis))+ ' años bisiestos')
    else:
        print('No hay años bisiestos entre los años entregados')
bisiestos(primero,segundo)
```
